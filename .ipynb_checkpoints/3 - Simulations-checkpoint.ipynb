```python
def sim_state(s):
    '''
    Pass in individual state s, and we simulate for each expected voter in the state.
    
    Returns the a list with [state, and the number of votes for biden, trump, and undecided (other)]
    '''
    
    tempState = states[states['state'] == s]
    
    biden = round(tempState['biden_avg'].values[0], 2)/100
    trump = round(tempState['trump_avg'].values[0], 2)/100
    voters = tempState['voting_pop'].values[0]
    reg_percent = tempState['percent_reg'].values[0]

    choices = ['biden', 'trump', 'undecided']

    und = (1 - biden - trump)
    probabilities =  [biden, trump, und]
    vals = np.random.choice(choices, voters, p = probabilities)
    biden_votes = np.sum(vals == "biden")
    trump_votes = np.sum(vals == "trump")
    undecided = np.sum(vals == "undecided")
    
    
    return([s, biden_votes, trump_votes, undecided])
```

```python
def simulate_election(polled_states, writer):
    
    '''
    Pass in list of states we want to simulate, and we simulate for each expected voter in the state.
    
    Returns the a df with column names ['state', 'biden', 'trump', 'undecided'] for the list of states simulated.
    '''
    column_names = ['state', 'biden', 'trump', 'undecided']
    election_df = pd.DataFrame(columns = column_names)
    for i in polled_states:
        election_value = sim_state(i)
        #write to csv
        writer.writerow(election_value)
        
        x = dict((val, election_value[idx]) for idx, val in enumerate(column_names))
        election_df = election_df.append(x, ignore_index=True)
        election_df.append(election_value)
    return (election_df)
```

```python
# reading in our Data

electoral_votes = pd.read_csv("Project Reveal/data/Electoral/electoral_votes.csv")
registered_voters = pd.read_csv('Project Reveal/data/USA/registered_voters_by_state.csv')
electoral_forecasts = pd.read_csv('Project Reveal/data/Polling/Electoral Forecasts.csv')
pres_return_2016 = pd.read_csv('Project Reveal/data/Returns/2016_pres_returns.csv')
```

```python

sanitized_df = pp.sanitize_Polls()

swing_states = ['Alaska', 'North Carolina', 'Florida', 'Arizona', 
                'Texas', 'Georgia', 'Ohio', 'Missouri', 'Arkansas', 'Iowa', 'Nevada']
polled_states = sanitized_df['state'].unique()
all_states = registered_voters['State'].unique()
```

```python


safe_red = electoral_forecasts[electoral_forecasts['Consensus (July)']>=6.1]["State"].values
safe_blue = electoral_forecasts[electoral_forecasts['Consensus (July)']<=2.1]["State"].values
safe_blue = np.append(safe_blue, "Maine")

column_names = ['state', 'biden_avg', 'trump_avg', 'voting_pop', 'percent_reg']
states = pd.DataFrame(columns = column_names)

for s in all_states:
    temp_state = sanitized_df[(sanitized_df["state"]==s)]
    reg_voters = registered_voters[registered_voters['State'] == s]
    
    if (s in safe_blue):
            safe_state = pres_return_2016[pres_return_2016['State']==s]
            state_values = [s, round(safe_state.Dem_Pct.values[0]*100, 3),
                    round(safe_state.Rep_Pct.values[0]*100, 3), 
                    reg_voters.totalRegistered.values[0], 
                    reg_voters.registeredPerc.values[0] 
                   ]
    elif (s in safe_red):
            safe_state = pres_return_2016[pres_return_2016['State']==s]
            state_values = [s, round(safe_state.Dem_Pct.values[0]*100, 3),
                    round(safe_state.Rep_Pct.values[0]*100, 3), 
                    reg_voters.totalRegistered.values[0], 
                    reg_voters.registeredPerc.values[0] 
                   ]
    else:
            state_values = [s, round(temp_state.biden_pct.mean(), 3),
                    round(temp_state.trump_pct.mean(), 3), 
                    reg_voters.totalRegistered.values[0], 
                    reg_voters.registeredPerc.values[0] 
                   ]
    x = dict ((val, state_values[idx]) for idx, val in enumerate(column_names))
    states = states.append(x, ignore_index=True)

    
# adding a few more columns for analysis later.    
states['eligible_pop'] = states['voting_pop']/states['percent_reg']
states['% diff'] = states['biden_avg'] - states['trump_avg']
```

```python
def sim_x(x):
    for i in range(x):
        with open('simulated_elections.csv','a', newline='') as f:
            writer = csv.writer(f)
            simulate_election(all_states, writer)
```
